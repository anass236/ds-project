# Imputation 

## Linear imputation

```python
AB_NYC_Linear = AB_NYC_2019.copy(deep = True)
```

```python
AB_NYC_Linear_review = AB_NYC_Linear[['price', 'number_of_reviews']]
```

```python
AB_NYC_Linear_review
```

```python
AB_NYC_Linear_review.interpolate(method = "linear",inplace = True)
```

```python
AB_NYC_Linear_review
```

```python
AB_NYC_Linear_review.describe()
```

```python
AB_NYC_Linear.number_of_reviews.describe()
```

```python
plt.scatter(AB_NYC_Linear_review.price,AB_NYC_Linear_review.number_of_reviews)
plt.xlabel("prix")
plt.ylabel("number_of_views")
plt.scatter(AB_NYC_Linear.price,AB_NYC_Linear.number_of_reviews,color = 'r')
```

## Quadratic imputation

```python
AB_NYC_quadratic_review = AB_NYC_Linear[['price', 'number_of_reviews']]
```

```python
AB_NYC_quadratic_review.interpolate(method='quadratic',inplace = True)
```
