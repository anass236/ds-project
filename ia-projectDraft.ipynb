## Preprocessing

```python
airbnb = AB_NYC_2019.copy()
airbnb.drop(columns=["name", "host_name","last_review"],inplace=True)
airbnb.reviews_per_month = airbnb.reviews_per_month.replace(np.nan,0)
airbnb.number_of_reviews = airbnb.number_of_reviews.replace(np.nan,0)
```

```python
airbnb.neighbourhood_group = airbnb.neighbourhood_group.factorize()[0]
airbnb.neighbourhood = airbnb.neighbourhood.factorize()[0]
airbnb.room_type = airbnb.room_type.factorize()[0]
airbnb.sort_values(by='price',ascending=True,inplace=True)
airbnb.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
airbnb.dropna(inplace=True)
```

```python
response = 'price'
features = [col for col in airbnb if col != response]
airbnb_500 = airbnb[airbnb.price< 500]
X_train, X_test, y_train, y_test = train_test_split(
    airbnb_500[features],
    airbnb_500[response],
    test_size=0.25
)
```

```python
model = LinearRegression(copy_X= True, fit_intercept = True, normalize = True)
model.fit(X_train, y_train)
```

```python
predicts = model.predict(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error

print("Root mean squared error is:")
print(np.sqrt(metrics.mean_squared_error(y_test,predicts)))
print('r2 score is:')
r2 = r2_score(y_test,predicts)
print(r2*100)
```

```python
error_airbnb = pd.DataFrame({
        'Actual Values': np.array(y_test).flatten(),
        'Predicted Values': predicts.flatten()})

error_airbnb
```
